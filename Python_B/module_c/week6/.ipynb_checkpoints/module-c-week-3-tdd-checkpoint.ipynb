```python
'''
The program shall open a named file.
The program shall be able to open a named file in read-only mode.
The program shall be able to open a named file in read-write mode.
'''

FILENAME = "carroll1.txt"

# UTILITY CODE - some functions we might need for every test
def cleanup_after_test(test_file):
    try:
        test_file.close()
    except Exception as ignored:
        pass

# FUNCTION STUBS - we'll complete these once we have written a test
def open_file (filename, mode='r'):
    return open(filename, mode)

def open_file_read_only(filename, mode='r'):
    f = open(filename, mode)
    return f

def open_file_read_write(filename, mode='a+'):
    f = open(filename, mode)
    f.seek(0, 0)
    return f

# UNIT TESTS - we'll complete these before completing the stubs above
def test_open_file ():
    
    # Your code here...
    test_file = open_file(FILENAME)
    
    # Make sure the file is really open by maybe reading a line from the file
    # although you can be creative here - designing the test is an important 
    # element of TDD
    
    test_data = test_file.read(5)

    # Prove your requirement or throw an AssertionError with a message
    assert len(test_data) == 5, "test_open_file failed"

    # And call this at the conclusion of each test to ensure the file
    # is closed ready for the next test.
    cleanup_after_test(test_file)

    # At the end of your test, it's good practice to return True - but optional.

    
def test_open_file_read_only ():

    # Call open_file_read_only
    test_file = open_file_read_only(FILENAME)
    
    # Test reading data
    test_data = test_file.read(5)
    
    assert len(test_data) == 5, "Test open file failed"
    
    # Test writing data, if it can write data then test fail
    try:
        test_file.write("Testing")
        file = open(FILENAME)
        test_data_after_write = file.read(5)
        file.close()
    
        assert test_data == test_data_after_write, "Test read only fail"
    except:
        pass
    
    cleanup_after_test(test_file)

def test_open_file_read_write ():

    # Your code here...
    test_file = open_file_read_write(FILENAME)
    
    # Test reading data
    test_data = test_file.read(5)

    assert len(test_data) == 5, "Test open file failed"
    
    # Test writing data, if it can write data then test pass
    test_file.write("Testing")
    test_file.seek(0,0)
    test_data_after_write = test_file.read()
    
    assert "Testing" in test_data_after_write, "Test write fail"

    cleanup_after_test(test_file)
    
# TEST SUITE

# initally, we test only the first requirement
test_open_file()

# these tests we will enable later

test_open_file_read_only ()
test_open_file_read_write ()

```
